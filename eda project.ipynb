## Questions About Data:
Ques- WHat are the locations with the fastest roller coasters (minimum of 10)?


```python
df["Location"].value_counts()
```

```python
a=df.query("Location!='Other'")\
.groupby("Location")["Speed_mph"].agg(['mean', 'count'])\
.query('count>=10')\
.sort_values("mean")
a
```

```python
a['mean'].plot(kind="barh", title="Average Coast speed by Loaction")
plt.xlabel("avg coaster speed")
plt.ylabel('Locations')
plt.show()
```
